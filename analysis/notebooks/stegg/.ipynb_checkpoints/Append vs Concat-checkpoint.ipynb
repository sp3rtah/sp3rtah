```python
from time import sleep, process_time_ns

length = int(1e5)
```

## String +=

```python
def str_concat():
    out = ''
    start = process_time_ns()
    for i in range(length):
        out += 'A'
    return process_time_ns() - start
    
```

## List append

```python
def list_append():
    out = list()
    start = process_time_ns()
    for i in range(length):
        out.append('A')
    return process_time_ns() - start
```

# Run and Compare

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
runs = 42
strings = np.array([str_concat() for _ in range(runs)])
lists = np.array([list_append() for _ in range(runs)])
```

```python
fig, (p1,p2) = plt.subplots(1,2)

p1.set_title('String Concat')
p1.plot(strings)

p2.set_title('List Append')
p2.plot(lists)
```

```python
plt.pie([lists.mean(),strings.mean()],labels=['List append','String concat'])
```
